## 1.1 Create a two-dimensional vector and plot it on a graph

```
import math
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

vec_2d = [5, 7]

plt.arrow(0,0, vec_2d[0], vec_2d[1], head_width = .5, head_length = .5, color = 'red')
plt.xlim(-1,11)
plt.ylim(-1, 11)
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
vec_3d = np.array([[0, 0, 0, 5, 7, 10]])

X, Y, Z, U, V, W = zip(*vec_3d)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
